```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time
import pandas

# Creates a csv file and writes the first row
csv_file = open("binghamton_job_openings.csv", "w", newline= "", encoding= "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Job", "Type", "Application Deadline", "Salary"])


# Allows communication between jupyter notebook and the browser
PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service= PATH)

# Opens the url in the browser
driver.get("https://www.binghamton-ny.gov")

# These XPaths are equal to the links to get to the page containing the list of jobs
government_tab = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/a")
departments = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/div/div/ul[1]/li/a")

# This action chain goes from the home page to the page with the list of jobs
actions = ActionChains(driver)
actions.move_to_element(government_tab)
actions.click(government_tab)
actions.click(departments)
actions.perform()

try:
#     Waits for the presence of the given XPath and then clicks on it once it shows up on the page
    personell_civil_service = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='leftNav_2_0_127']/ul/li/ul/li[14]/a[1]"))
    )
    personell_civil_service.click()
    
#     Waits for the presence of the given XPath and then clicks on it once it shows up on the page
    employment = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='leftNav_1038_0_145']/ul/li/ul/li[14]/ul/li/a"))
    )
    employment.click()
    
#     Establishes the number of rows in the job list for the for loop
    num_rows = len(driver.find_elements(By.XPATH, "//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr"))
    
#     Runs through each row and obtains the following information from each column in each row by modifying the XPath
    for row in range(1, num_rows + 1):
        job_title = driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{row}]/td[1]").text
        job_type = driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{row}]/td[2]").text
        application_deadline = driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{row}]/td[3]").text
        salary = driver.find_element(By.XPATH, f"//*[@id='ColumnUserControl4']/div[2]/table/tbody/tr[{row}]/td[4]").text
        csv_writer.writerow([job_title, job_type, application_deadline, salary])

finally:
#     After that code has run it will sleep for 5 seconds then close the browser
    time.sleep(5)
    driver.quit()
    
#     Closes the csv file
    csv_file.close()
    
#     Makes the csv file more readable and prints it
    pand = pandas.read_csv('binghamton_job_openings.csv')
    print(pand.to_string())
```
