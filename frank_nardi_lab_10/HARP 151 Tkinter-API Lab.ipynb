```python
from tkinter import *
import json
import requests
import random

# Gets a list of 5 random authors from the quotes API
# Get authors has been working fine for a long time and suddenly stopped, I keep getting "JSONDecodeError: Expecting value: line 1 column 1 (char 0)" I'm not sure what's going wrong
def get_authors():
    author_list = []
    for _ in range(5):
        page_num = random.randint(1, 41)
        p = requests.get(f"https://api.quotable.io/authors?sortBy=name&page={page_num}")
        page = p.json()
        auth_amt = len(page['results'])
        author_index_on_page = random.randint(0, auth_amt - 1)
        author_name = page['results'][author_index_on_page]['name']
        author_list.append(author_name)
    return(author_list)
    
def get_quote(auth):
    r = requests.get(f"https://api.quotable.io/random?author={auth}")
    quote_info = r.json()
    quote = quote_info['content']
    return(quote)

def show(x):
    value.set(x)

def reset_list():
    default = "Select a Person"
    value.set(default)
    new_list = get_authors()
    menu = dropdown["menu"]
    menu.delete(0, END)
    for auth in new_list:
        menu.add_command(label= auth, command= lambda x= auth: show(x))


def chosen_option():
    output_text.delete(0.0, END)
    author = dropdown.cget("text")
#     if author != "Select a Person":
    quote = get_quote(author)
    output_text.insert(END, f"Your quote is: \"{quote}\"")
    
item_list = get_authors()

root = Tk()
root.geometry("400x400")

value = StringVar()
value.set("Select a Person")

dropdown = OptionMenu(root, value, *item_list)
choose_button = Button(root, text= "Click to choose", command= chosen_option)
new_list = Button(root, text= "Click to reset list", command= reset_list)
output_text = Text(root, height= 4, width= 30)

dropdown.pack()
choose_button.pack()
new_list.pack()
output_text.pack()

root.mainloop()
```

```python
from tkinter import *
import json
import requests

# I tried using PanedWindow to make it look a little nicer, but couldn't really figure it out
# Here's the code I tried using to do that, which replaces some code towards the bottom of this:
# name_label = Label(root)
# name_label.pack()
# frame = Frame(root)
# test = PanedWindow(root, bg= "lightblue")

# for i in range(len(forecast_list)):
#     test.add(frame, text= f"{forecast_list[i][0]} \n {forecast_list[i][1]}")


root = Tk()
root.geometry("500x100")
lat = "40.741895"
lon = "-73.989308"
forecast_list = []

def get_forecast(lat, lon):
    genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
    json_file = genius.json()

    forecast = json_file['properties']['forecast']
    new_request = requests.get(forecast)

    json = new_request.json()

    weather_list = json['properties']['periods']

    for i in range(len(weather_list)):
        name = json['properties']['periods'][i]['name']
        temp = json['properties']['periods'][i]['temperature']
        tup = (name, temp)
        if not "Night" in tup[0]:
            forecast_list.append(tup)

    title = Label(root, text= "Weather Forecast", font= "Arial")
    title.pack()
    name_label = Label(root)
    name_label.pack()
    frame = Frame(root)
    frame.pack()
    frame.pack_propagate(0)

    for i in range(len(forecast_list)):
        day = Label(frame, text= f"{forecast_list[i][0]} \n {forecast_list[i][1]}")
        day.grid(row= 0, column= i)

    root.mainloop()
    
get_forecast(lat, lon)
```
