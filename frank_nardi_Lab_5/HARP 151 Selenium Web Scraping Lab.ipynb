```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv

# This line opens the csv or creates one if one with the given name does not exist
# The following creates the thing that will edit the csv
# The last of the three writes a title for the row ("Image URL")
csv_file = open("wikipedia_cats.csv", "w", newline= "", encoding= "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Image URL"])

# This is meant to serve as a reference so the code knows which browser service to use with Selenium
PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service= PATH)

# Here the driver takes opens the given URL
driver.get("https://www.wikipedia.org/")

# This code establishes 'search' as the searchbar which is found by its ID
# The searchbar is cleared in case of any text may be cluttering it
# The word 'cat' is typed into it using 'send_keys'
# Also using 'send_keys', 'enter' is input
search = driver.find_element(By.ID, "searchInput")
search.clear()
search.send_keys("cat")
search.send_keys(Keys.RETURN)

# The code searches for all elements with the 'img' tag and puts them into the list 'images'
# The for loop then is able to account for all of the images on the page and takes their url using 'get_attribute'
# I printed the url to ensure that they were correct and all present in accordance with the html code from the wikipedia page
# Then the url are all added to the csv using 'csv_writer'
# After this is completed the driver is closed using 'quit'
try:
    images = driver.find_elements(By.TAG_NAME, "img")
    
    for i in images:
        url = i.get_attribute("src")
#         print(url)
        csv_writer.writerow([url])
        
finally:
    driver.quit()
```
